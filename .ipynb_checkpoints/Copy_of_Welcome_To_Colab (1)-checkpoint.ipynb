It looks like the target is of qualitative ordinal type because each class is shown in order of severity. The treatment of this variable will be done by transforming each class to its respective severity number. The more serious the accident the higher number it will get.

```python
# Mapeamento de cada valor categórico para um número, incluindo o prefixo numérico
injury_type_mapping = {
    '2. NON-COMP': 2,
    '4. TEMPORARY': 4,
    '3. MED ONLY': 3,
    '5. PPD SCH LOSS': 5,
    '6. PPD NSL': 6,
    '1. CANCELLED': 1,
    '8. DEATH': 8,
    '7. PTD': 7,
    'Não especificado': 0  # Valor para representar os NaN #-exprimentar dropar as linhas onde é igual a 0?
}

# Substituir valores NaN por "Não especificado" antes de mapear
train_df["Claim Injury Type"] = train_df["Claim Injury Type"].fillna("Não especificado")

# Remover espaços em branco ao redor dos valores, caso necessário
train_df["Claim Injury Type"] = train_df["Claim Injury Type"].str.strip()

# Aplicar o mapeamento atualizado à coluna "Claim Injury Type"
train_df["Claim Injury Type"] = train_df["Claim Injury Type"].map(injury_type_mapping)

# Exibir os valores únicos para verificar o resultado
print(train_df["Claim Injury Type"].unique())
```

# Dealing with the rest of the missing values

There are some missing values that we can build an aproximation to their real values, for example 'Birth Year'. The values of this column are dependent on the 'Accident date' and 'Age at Injury'.

```python
# Conta o número de valores ausentes nas colunas específicas
missing_values_selected = train_df[["Accident Date", "Age at Injury", "Birth Year"]].isna().sum()

# Exibe o número de valores ausentes em cada coluna selecionada
print("Número de valores ausentes nas colunas 'Accident Date', 'Age at Injury' e 'Birth Year':")
print(missing_values_selected)
```

First lets check if there is any inconsistency with the 'Accident Date' date formats.

```python
#Check the accident date column for inconsistency
# Check the unique formats in the column as strings
date_formats = train_df["Accident Date"].dropna().astype(str).str.match(r'^\d{4}-\d{2}-\d{2}$')
if date_formats.all():
    print("All dates in 'Accident Date' are in YYYY-MM-DD format.")
else:
    print("Some dates in 'Accident Date' do not match the YYYY-MM-DD format.")
```

The approximations for 'Birth Year' column missing values will be done using the following code.

```python
#Every Accident Date is in YYYY-MM-DD format and we have 3689 missing values. Also, is the mouth and day that important? maybe we could drop it because we dont have the same format on birth year.
#Lets try to fill some missing values on Birth Year:

# Ensure "Accident Date" is in datetime format
train_df["Accident Date"] = pd.to_datetime(train_df["Accident Date"], errors='coerce')

# Calculate "Birth Year" for missing values using "Accident Date" and "Age at Injury"
train_df["Birth Year"] = train_df.apply(
    lambda row: row["Accident Date"].year - row["Age at Injury"] if pd.isna(row["Birth Year"]) and pd.notna(row["Accident Date"]) and pd.notna(row["Age at Injury"]) else row["Birth Year"],
    axis=1
)
```

```python
# Conta o número de valores ausentes nas colunas específicas
missing_values_selected = train_df[["Accident Date", "Age at Injury", "Birth Year"]].isna().sum()

# Exibe o número de valores ausentes em cada coluna selecionada
print("Número de valores ausentes nas colunas 'Accident Date', 'Age at Injury' e 'Birth Year':")
print(missing_values_selected)
```

Using the same logic to calculate the Accident date

```python
# Preencher valores ausentes em "Accident Date" usando "Birth Year" e "Age at Injury"
train_df["Accident Date"] = train_df.apply(
    lambda row: row["Birth Year"] + row["Age at Injury"] if pd.isna(row["Accident Date"]) and pd.notna(row["Birth Year"]) and pd.notna(row["Age at Injury"]) else row["Accident Date"],
    axis=1
)
```

# Feature Selection

The dataset for the predictions will be created based on 'train_df'. Some variables need to be separated as dummies for the model to interpret them correctly

```python
# Dataframe para usar para a modelação
train_df_filtered = train_df

# Apagar as colunas que não vamos precisar ou que já temos acesso através dos códigos
#train_df_filtered = train_df_filtered.drop(columns=['Claim Identifier'])
train_df_filtered = train_df_filtered.drop(columns=['Industry Code Description'])
train_df_filtered = train_df_filtered.drop(columns=['WCIO Nature of Injury Description'])
train_df_filtered = train_df_filtered.drop(columns=['WCIO Cause of Injury Description'])
train_df_filtered = train_df_filtered.drop(columns=['WCIO Part Of Body Description'])
#train_df_filtered = train_df_filtered.drop(columns=['Agreement Reached'])
train_df_filtered = train_df_filtered.drop(columns=['WCB Decision'])

# Separar os dados das datas em Ano, Mês e Dia e apagar a coluna com a data inteira

train_df_filtered['Accident_Year'] = pd.to_datetime(train_df_filtered['Accident Date']).dt.year
train_df_filtered['Accident_Month'] = pd.to_datetime(train_df_filtered['Accident Date']).dt.month
train_df_filtered['Accident_Day'] = pd.to_datetime(train_df_filtered['Accident Date']).dt.day
train_df_filtered = train_df_filtered.drop(columns=['Accident Date'])

train_df_filtered['Assembly_Year'] = pd.to_datetime(train_df_filtered['Assembly Date']).dt.year
train_df_filtered['Assembly_Month'] = pd.to_datetime(train_df_filtered['Assembly Date']).dt.month
train_df_filtered['Assembly_Day'] = pd.to_datetime(train_df_filtered['Assembly Date']).dt.day
train_df_filtered = train_df_filtered.drop(columns=['Assembly Date'])

train_df_filtered['C-2_Year'] = pd.to_datetime(train_df_filtered['C-2 Date']).dt.year
train_df_filtered['C-2_Month'] = pd.to_datetime(train_df_filtered['C-2 Date']).dt.month
train_df_filtered['C-2_Day'] = pd.to_datetime(train_df_filtered['C-2 Date']).dt.day
train_df_filtered = train_df_filtered.drop(columns=['C-2 Date'])

train_df_filtered['C-3_Year'] = pd.to_datetime(train_df_filtered['C-3 Date']).dt.year
train_df_filtered['C-3_Month'] = pd.to_datetime(train_df_filtered['C-3 Date']).dt.month
train_df_filtered['C-3_Day'] = pd.to_datetime(train_df_filtered['C-3 Date']).dt.day
train_df_filtered = train_df_filtered.drop(columns=['C-3 Date'])

train_df_filtered['First_Hearing_Year'] = pd.to_datetime(train_df_filtered['First Hearing Date']).dt.year
train_df_filtered['First_Hearing_Month'] = pd.to_datetime(train_df_filtered['First Hearing Date']).dt.month
train_df_filtered['First_Hearing_Day'] = pd.to_datetime(train_df_filtered['First Hearing Date']).dt.day
train_df_filtered = train_df_filtered.drop(columns=['First Hearing Date'])

# variables with lots of dummies
# 'Carrier Name', 'County of Injury', 'WCIO Cause of Injury Code', 'WCIO Nature of Injury Code', 'WCIO Part Of Body Code', 'Zip Code'
# Nota: Ainda podemos incluir estas variáveis no modelo, mas como têm muitas categorias talvez tentar fazer grupos

# dropping high variables with many dummies for now
train_df_filtered = train_df_filtered.drop(columns=['Carrier Name'])
train_df_filtered = train_df_filtered.drop(columns=['County of Injury'])
train_df_filtered = train_df_filtered.drop(columns=['WCIO Cause of Injury Code'])
#train_df_filtered = train_df_filtered.drop(columns=['WCIO Nature of Injury Code'])
#train_df_filtered = train_df_filtered.drop(columns=['WCIO Part Of Body Code'])
train_df_filtered = train_df_filtered.drop(columns=['Zip Code'])

#geting the dummies
dummy_columns = ['Alternative Dispute Resolution', 'Attorney/Representative', 'Carrier Type',
    'COVID-19 Indicator', 'District Name', 'Gender', 'Industry Code', 'Medical Fee Region',
                 'Agreement Reached', 'WCIO Part Of Body Code', 'WCIO Part Of Body Code']
train_df_filtered = pd.get_dummies(train_df_filtered, columns=dummy_columns)
```

```python
# Is 'IME-4 Count' and 'Number of Dependents' ordinal or should they be treated as dummies?
```

```python
train_df['Agreement Reached'].value_counts()
```

The dummie columns values are substituted by 1 if True or 0 if False for the model to use them properly.

```python
train_df_filtered = train_df_filtered.replace({True: 1, False: 0})
```

The columns of the dataset are then grouped in X if independent or y if dependent

```python
X = train_df_filtered.drop('Claim Injury Type', axis=1)  # Replace 'Claim Injury Type' with your target variable
y = train_df_filtered['Claim Injury Type']
```

Then we take a subset for training and one for validation and do the same separation between dependent and independent types of variable. Those are all subsets of the training dataset.

```python
X_train, X_val, y_train, y_val = train_test_split(X,y, test_size = 0.2,
                                                  stratify = y,
                                                  shuffle = True)
```

The missing values of the independent vaiables are filled using the mean in both the X_train and X_val datasets

```python
#filling with mean
for column in X_train.columns:
    if pd.api.types.is_numeric_dtype(X_train[column]):

        #store mean of training data in a variable - in a real application, you may need to store these values for future usages on e.g. test data
        mean_to_fill = X_train[column].mean()

        #fill on X_train
        X_train[column].fillna(mean_to_fill, inplace=True)
        #Fill on X_val
        X_val[column].fillna(mean_to_fill, inplace=True)
```

The independent variables need to be scaled to facilitate processing and better comparison between variables

```python
# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Select only the numeric columns for normalization
numeric_cols = X_train.select_dtypes(include=['number']).columns

# Fit the scaler on the training data and transform both training and validation data
X_train[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])
X_val[numeric_cols] = scaler.transform(X_val[numeric_cols])
```

```python
model = LogisticRegression()
rfe = RFE(estimator = model, n_features_to_select = 40)

X_rfe = rfe.fit_transform(X = X_train, y = y_train)
```

```python
# Get the selected features from RFE
selected_features = X_train.columns[rfe.support_].tolist()

# Print the selected features
print("Selected Features:")
selected_features
```

```python
X_train = X_train[selected_features]
X_val = X_val[selected_features]
```

```python
# prompt: Build a simple logistic regression model based on those features selected in the last code snipet
model = LogisticRegression()  # You can adjust hyperparameters as needed
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_val)

# Evaluate your model

accuracy = accuracy_score(y_val, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# Step 3: Initialize the Random Forest model
model = RandomForestClassifier()

# Step 4: Train the model
model.fit(X_train, y_train)

# Step 5: Make predictions on the validation set
y_pred = model.predict(X_val)

# Step 6: Evaluate the model
print("Accuracy:", accuracy_score(y_val, y_pred))
```
